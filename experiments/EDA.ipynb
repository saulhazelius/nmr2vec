### Simple Data Exploration

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

### Read data file 


```python
def read_data(file):
    """Returns a np array from file. The resultant dimensions are N lines X variable size."""

    X = []
    y = []
    for line in file:
        spectra = [float(nmr) for nmr in line.split(", ")[:-1]]
        X.append(np.array(spectra))
        y.append(np.array(int(line.split(", ")[-1].strip())))

    X = np.array(X, dtype=object)
    y = np.array(y)

    return (X, y)
```

```python
file = open("../data/steroids.csv")
```

```python
X, y = read_data(file)
```

### We are gonna analize the NMR spectra corresponding to the steroids (y = 1)

```python
# Get postive idcs:
pos_idcs = (y == 1)
```

```python
X_pos = X[pos_idcs] # Steroids NMR
```

```python
# X_pos is an array of variable size arrays. We have to convert it to another iterable for performing a frequency dist plot
X_pos_li = []
for array in X_pos:
    X_pos_li.extend(array)
```

```python
sns.displot(X_pos_li, kind="kde")
```

### Same for the non-steroid NMR (y = 0)

```python
neg_idcs = (y == 0)
```

```python
X_neg= X[neg_idcs] # Steroids NMR
```

```python
X_neg_li = []
for array in X_pos:
    X_neg_li.extend(array)
```

```python
sns.displot(X_neg_li, kind="kde")
```
